## Define Model

```python
class simplernn(nn.Module):
  def __init__(self, input_size, hidden_size, output_size):
    super(simplernn, self).__init__()
    self.rnn = nn.RNN(input_size, hidden_size, batch_first = True)
    self.fc = nn.Linear(hidden_size, output_size)

  def forward(self, x):
    out, hidd = self.rnn(x)
    final_hidden_state = hidd[-1]
    return self.fc(final_hidden_state)
```

```python
inp = 5
hid = 3
out = 2

model = simplernn(inp, hid, out)
```

## Training

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.01)
```

```python
#zerograd,forward,loss,backward,update
for epoch in range(3):
  running_loss = 0
  model.train()
  output = model(X)
  loss = criterion(output, y)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  running_loss += loss.item()
  print(f"Running Loss: {running_loss} for Epoch: {epoch}")
```

```python
def text_to_tensor(text):
  tokens = tokenizer(text)
  return torch.tensor(vocab(tokens))
```

```python
def collate_fn(batch):
  labels, texts = zip(*batch)
  texts = [text_to_tensor(text) for text in texts]
  texts = pad_sequence(texts, batch_first = True, padding_value = vocab['<pad>'])
  labels = torch.tensor([label for label in labels])
  return labels, texts
```

```python
from torch.utils.data import DataLoader
train_iter = datasets.IMDB(split = 'train')
```

```python
train_loader = DataLoader(train_iter,batch_size = 32, collate_fn = collate_fn)
```
